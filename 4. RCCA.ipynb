# One-Hold Regularized CCA
Splits the data in train (80%) and test (20%) at random and returns the canonical correlations for training and test data.

```r
RCCA_one_fold = function(X, Y, lambda1, lambda2 = 0){
    X = as.matrix(X)
    Y = as.matrix(Y)
    n = nrow(X)
    ind_train = sample(n, size = round(n * 0.8))
    
    X_train = X[ind_train, ]
    Y_train = Y[ind_train, ]
    X_test = X[-ind_train, ]
    Y_test = Y[-ind_train, ]
  
    RCC = rcc(X_train, Y_train, lambda1, lambda2)
    alpha = RCC$xcoef
    beta = RCC$ycoef
  
    cor_train = RCC$cor
    
    X_hat_test = X_test%*%alpha
    Y_hat_test = Y_test%*%beta
    cor_test = diag(cor(X_hat_test, Y_hat_test))
    
    return(list(cor_train = cor_train, cor_test = cor_test))
}
```

# Cross-Validated RCCA

Calculate 1SD confidence intervals for train and test canonical correlations.

```r
RCCA_CV = function(X, Y, lambda1, lambda2 = 0, nreps = 100){
    ncomp = min(ncol(X), ncol(Y))
    cor_train = matrix(0, nreps, ncomp)
    cor_test = matrix(0, nreps, ncomp)
    
    for(i in 1:nreps){
        model = RCCA_one_fold(X, Y, lambda1, lambda2)
        cor_train[i,] = model$cor_train
        cor_test[i,] = model$cor_test
    }
    
    #1-SD confidence intervals
    center_train = colMeans(cor_train)
    radius_train = colSds(cor_train)
    center_test = colMeans(cor_test)
    radius_test = colSds(cor_test)
    
    #plot CCA correlation plot
    p = ggplot(NULL, aes(x = 1:ncomp, y = center_train)) + 
        geom_point(size = 1, shape = 16, colour = 'red') +
        geom_errorbar(aes(ymin = center_train - radius_train, ymax = center_train + radius_train), colour = "black", width = .1) +
        geom_point(aes(x = 1:ncomp, y = center_test), size = 1, shape = 16, colour = 'blue') +
        geom_errorbar(aes(ymin = center_test - radius_test, ymax = center_test + radius_test), colour = "black", width = .1)+
        geom_hline(yintercept = 0, linetype = "dashed") +
        xlab('component') +
        ylab('correlation')+
        ggtitle(bquote(paste(lambda[1], '=', .(lambda1))))+
        theme(plot.title = element_text(size = 10))
        
    return(list(center = center_test[1], radius = radius_test[1], plot = p))
}
```

# Grid search for RCCA

Varying $\lambda_1$ and plot RCCA results for different values.

```r
lambda1s = 10^seq(-2,6)
nreps = 100
```

```r
nlam = length(lambda1s)
p = list()
centers = c()
radii = c()
for(i in 1:nlam){
    CI = RCCA_CV(c, q, lambda1s[i], 0, nreps)
    p[[i]] = CI$plot
    centers = append(centers, CI$center)
    radii = append(radii, CI$radius)
}
write.csv(data.frame('lambda1' = lambda1s, 'cor' = centers), file = paste('Data/Connectome/', c_folder_name, '/RCCA/rcca_cor.csv', sep = ''), row.names = FALSE)
ps = do.call("grid.arrange", c(p, ncol = 3))
ggsave(file = paste('Data/Connectome/', c_folder_name, '/RCCA/rcca_grid_search.png', sep = ''), device = 'png', plot = ps, width = 10, height = 16)
```

Search for the best $\lambda_1$ with the highest value of the first test canonical correlation

```r
lambda1_opt = lambda1s[which.max(centers)]
grid_search = ggplot(NULL, aes(x = log(lambda1s, 10), y = centers)) + 
        geom_point(size = 1, shape = 16, colour = 'blue') +
        geom_line(colour = 'blue') + 
        geom_errorbar(aes(ymin = centers - radii, ymax = centers + radii), colour = "black", width = .1) +
        geom_hline(yintercept = 0, linetype = "dashed") +
        xlab(bquote(paste('log(', lambda, ')', sep = ''))) +
        ylab('first canonical correlation')+
        ggtitle(bquote(paste('max correlation = ', .(round(max(centers), 2)), '     ', lambda, ' optimal = ', .(lambda1_opt), sep = ' ')))
grid_search
ggsave(file = paste('Data/Connectome/', c_folder_name, '/RCCA/rcca_best_lambda.png', sep = ''), device = 'png', plot = grid_search, width = 10, height = 5)
```

# Calculate RCCA for the best lambda.

```r
RCC = rcc(c, q, lambda1_opt, lambda2 = 0)
```

Save the loadings. 

```r
alpha = RCC$xcoef
colnames(alpha) = paste('RCCA', 1:ncol(alpha), sep = '')
beta = RCC$ycoef
colnames(beta) = paste('RCCA', 1:ncol(beta), sep = '')
write.csv(data.frame('PCA_comp' = rownames(alpha), alpha), file = paste('Data/Connectome/', c_folder_name, '/RCCA/alpha.csv', sep = ''), row.names = FALSE)
write.csv(data.frame('q' = rownames(beta), beta), file = paste('Data/Connectome/', c_folder_name, '/RCCA/beta.csv', sep = ''), row.names = FALSE)
```
